Running a cell means that you will execute the cell’s contents(=the computer will perform the logic of the code). To execute a cell, you can just select the cell and click the Run button that is in the row of buttons along the top. It’s towards the middle. If you prefer using your keyboard, you can just press Shift+Enter.

When you run a cell, you will notice that there are some square braces next to the word In to the left of the cell. The square braces will auto fill with a number that indicates the order that you ran the cells. For example, if you open a fresh Notebook and run the first cell at the top of the Notebook, the square braces will fill with the number 1.
When we run the cell, its output is displayed below and the label to its left will have changed from In [ ] to In [1].

mathematical priorities apply here (mulitplication has priority over addition):

```python
print(4 * 2 + 1) 
```

You can also use parentheses to force the priority

```python
print(4 * (2 + 1))
```

We can also store the result of a calculation into a new variable:

```python
c = a + b * a 
print(c)
```
